```python
import pandas as pd
```

```python
vine_df = pd.read_csv('vine_table.csv')
vine_df
```

#### 1. Filter for rows where Total Votes are greater than or equal to 20

```python
total_votes_20_or_more = (vine_df['total_votes'] >= 20)
vine_df_total_votes_filtered = vine_df.loc[total_votes_20_or_more]
vine_df_total_votes_filtered
```

#### 2. Filter the new DataFrame created in Step 1 and create a new DataFrame to retrieve all the rows where the number of helpful_votes divided by total_votes is equal to or greater than 50%.

```python
helpful_votes_filter = (vine_df_total_votes_filtered['helpful_votes']/vine_df_total_votes_filtered['total_votes'] >= 0.5)
vine_df_helpful = vine_df_total_votes_filtered.loc[helpful_votes_filter]
vine_df_helpful
```

#### 3. Filter the DataFrame or table created in Step 2, and create a new DataFrame or table that retrieves all the rows where a review was written as part of the Vine program (paid), vine == 'Y'.

```python
vine_program_filter = (vine_df_helpful['vine'] == 'Y')
vine_df_paid_review = vine_df_helpful.loc[vine_program_filter]
vine_df_paid_review
```

#### 4. Repeat Step 3, but this time retrieve all the rows where the review was not part of the Vine program (unpaid), vine == 'N'.

```python
vine_non_program_filter = (vine_df_helpful['vine'] == 'N')
vine_df_unpaid_review = vine_df_helpful.loc[vine_non_program_filter]
vine_df_unpaid_review
```
